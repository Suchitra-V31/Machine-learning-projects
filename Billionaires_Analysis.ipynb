In this dataset we are going to analyze the Billionaires data....Using some Visualization techniques.



```python
#import the libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

**Load the Data**

```python
data=pd.read_csv('/content/Billionaire.csv')
```

```python
data.head()
```

Let us get some information about the data.

```python
data.shape
```

```python
data.info()
```

Check whether the data has any nan values.


```python
sns.heatmap(data.isnull())
```

```python
data['Age']=data['Age'].fillna(data['Age'].mean()) # Replace the nan values with mean
```

Now we have removed all our nan values..We can see that there is no nan values present in our dataset.

```python
sns.heatmap(data.isnull()) 
```

Since Networth has some special characters , let us remove the all those characters and convert them into float.



```python
data['NetWorth']=data['NetWorth'].str.strip('$')
data['NetWorth']=data['NetWorth'].str.strip('B')
data['NetWorth']=data['NetWorth'].astype(float)
```

Let us visualize the Top 10 Billionaires with respect to their Networth.

```python
df = data.sort_values(by = ["NetWorth"], ascending=False).head(10)
plt.figure(figsize=(20, 10))
sns.histplot(x="Name", hue="NetWorth", data=df)
plt.show()
```

Let us visualize the Top domains to become a billionaire.

```python
a=data['Source'].value_counts().head()
```

```python
a
```

```python
index=a.index
source=a.values
color=['skyblue','pink','yellow','red','purple']
plt.figure(figsize=(5,5))
plt.pie(source,labels=index,colors=color)
plt.title("Top 5 Domains to Become a Billionaire", fontsize=20)
plt.show()
```

Here is a pie chart which depicts the Top industries with most no.of Billionaires.


```python
a=data['Industry'].value_counts().head(8)
```

```python
a
```

```python
index=a.index
industries=a.values
color=['#00876c','#5a9b76','#8aaf87','#b4c4a0','#d8d9be','#f6f1e1','#ead4b3','#e5b48b','#e1926b','#dc6b58','#d43d51']
plt.figure(figsize=(5,5))
plt.pie(industries,labels=index,colors=color)
central_circle=plt.Circle((0,0),0.5,color='white')
fig=plt.gcf()
fig.gca().add_artist(central_circle)
plt.title("Top 11 Industries with Most Number of Billionaires", fontsize=20)
plt.show()
```

Here is a barplot of countries which has most no of millionaires.
