Before we proceed let us import all the necessary libraries.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

Let us load the data.

```python
data=pd.read_csv('/content/Uber-Jan-Feb-FOIL.csv')
```

```python
data['date']=data['date'].map(pd.to_datetime)
```

Let us split the date column into Date,Day,Month,Year, so that it becomes easy for us to analyse.

```python
data['Date']=data['date'].apply(lambda x: x.day)
data['Day']=data['date'].apply(lambda x: x.weekday())
data['Month']=data['date'].apply(lambda x: x.month)
data['Year']=data['date'].apply(lambda x: x.year)
```

```python
data.head()
```

From this we could see that,

```python
data['Year'].unique()
```

```python
sns.set(rc={'figure.figsize':(12, 8)})
sns.distplot(data['Day'])
```

```python
sns.barplot(data['Day'],data['trips'])
```

From both of the graphs we could observe that during Wednesday - Friday uber has got many trips.

```python
sns.distplot(data['active_vehicles'])
```

From the above graph we could infer that the vehicles are active at the initial days but it got decreased suddenly and was active moderately for the remaining days.

```python
sns.barplot(data['Month'],data['active_vehicles'])
```

We could see that in the month of February the vehicles were active the most.

```python
sns.barplot(data['Month'],data['trips'])
```

Similarly we got more trips in the moth of February.

```python
sns.heatmap(data.corr(),annot=True)
```

We can see that the trips and active vehicles are correlated to each other. We can also see it in the scatter plot how they are correlated.

```python
sns.scatterplot(data['trips'],data['active_vehicles'])
```

```python
sns.distplot(data['Date'])
```

We could see that only at the month end we find the decrease in uber trips.
